```python
import polars as pl
import io
import requests

# Sample DataFrame
df = pl.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'value': [10.5, 20.3, 30.7]
})

# Serialize DataFrame to Parquet in memory
buffer = io.BytesIO()
df.write_parquet(buffer, compression='snappy')
buffer.seek(0)  # Reset buffer position to the beginning

# Send the bytes directly to the API
API_URL = 'http://localhost:8082/upload'
headers = {'X-API-Key': 'your-secure-api-key'}

# Use buffer.getvalue() to get the bytes data
files = {'file': ('data.parquet', buffer.getvalue(), 'application/octet-stream')}

response = requests.post(API_URL, files=files, headers=headers)
print(response.status_code)
print(response.json())
```
